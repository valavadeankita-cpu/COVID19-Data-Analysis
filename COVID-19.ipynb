```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Data Loading

```python
df=pd.read_csv('country_wise_latest.csv')
```

## Data Cleaning

```python
df=df.dropna(subset=['Country/Region','Confirmed','Deaths','Recovered','New cases','New deaths','New recovered','1 week % increase','Recovered / 100 Cases','Deaths / 100 Cases'])
```

### Top 10 Countries With Highest Confirmed Cases

```python
country_count = df.nlargest(10,'Confirmed')
plt.figure(figsize=(10,6))
plt.bar(country_count['Country/Region'],country_count['Confirmed'],color='orange')
plt.title('Top 10 countries with highest confirmed cases')
plt.xlabel('country')
plt.ylabel('Number of cases')
plt.ticklabel_format(style='plain',axis='y')
plt.tight_layout()
plt.savefig('top_10_highest_confirmed.png')
plt.show()
```

**"This bar chart shows top 10 countries with highest confirmed  COVID-19 cases.Countries like US and Brazil have significantly higher cases compare to others"**

### Top 10 Countries With Highest Deaths Cases

```python
country_count = df.nlargest(10,'Deaths')
plt.figure(figsize=(10,6))
plt.bar(country_count['Country/Region'],country_count['Deaths'],color='red')
plt.title('Top 10 countries with highest Deaths cases')
plt.xlabel('country')
plt.ylabel('Number of Deaths')
plt.ticklabel_format(style='plain',axis='y')
plt.tight_layout()
plt.savefig('top_10_highest_Deaths.png')
plt.show()
```

**"This bar chart shows top 10 countries with highest Deaths cases.it highlights which countries were most affected in terms of fatalities."**

### Top 10 Countries With Highest Recovered Cases

```python
country_count = df.nlargest(10,'Recovered')
plt.figure(figsize=(10,6))
plt.bar(country_count['Country/Region'],country_count['Recovered'],color='green')
plt.title('Top 10 countries with highest Recovered cases')
plt.xlabel('country')
plt.ylabel('Number of Recovered')
plt.ticklabel_format(style='plain',axis='y')
plt.tight_layout()
plt.savefig('top_10_highest_recovered.png')
plt.show()
```

**"This bar chart shows top 10 countries with highest Recovered cases.it indicates that countries with high cases also have high recoveries."**

### Distribution Of Confirmed Cases

```python
plt.figure(figsize=(10,6))
plt.hist(df['Confirmed'],bins=50,color='purple',edgecolor='black')
plt.title('distribution of confirmed cases')
plt.xlabel('confirmed cases')
plt.ylabel('number of countries')
plt.ticklabel_format(style='plain',axis='y')
plt.tight_layout()
plt.savefig('distribution_confirmed_cases.png')
plt.show()
```

**"This histogram shows how confirmed cases are distributed across countries.Most counteries have low cases,while few have very high cases"**

### Relationship Between Confirmed Cases And Deaths

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Confirmed'],df['Deaths'],color='blue')
plt.title('relation b/w confirmed cases and deaths')
plt.xlabel('confirmed cases')
plt.ticklabel_format(style='plain',axis='x')
plt.ylabel('deaths')
plt.ticklabel_format(style='plain',axis='y')
plt.tight_layout()
plt.savefig('confirmed_cases_and_deaths.png')
plt.show()
```

**"This scatter plot shows relationship between confirmed cases and deaths.counteries with higher cases tend to have higher deaths"**

### Countries With Highest New Cases

```python
plt.figure(figsize=(10,6))
plt.bar(country_count['Country/Region'],country_count['New cases'],color='darkred')
plt.title('countries with highest New cases')
plt.xlabel('country')
plt.ylabel('Number of New cases')
plt.ticklabel_format(style='plain',axis='y')
plt.tight_layout()
plt.savefig('top_10_highest_New_cases.png')
plt.show()
```

**"This bar chart shows top 10 counteries with highest new cases.it helps identify where the spread is currently increasing."**

### Countries With Highest 1 Week % Increase

```python
plt.figure(figsize=(10,6))
plt.bar(country_count['Country/Region'],country_count['1 week % increase'],color='teal')
plt.title('countries with highest 1 week %')
plt.xlabel('country')
plt.ylabel('Number of highest 1 week %')
plt.ticklabel_format(style='plain',axis='y')
plt.tight_layout()
plt.savefig('top_10_highest_1_week_%.png')
plt.show()
```

**"This bar chart shows top 10 counteries with highest weekly growth in cases.it indicates which counteries are experiencing rapid increase."**

### Compare Deaths VS Recoveries

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Deaths'],df['Recovered'],color='green')
plt.title('relation b/w  deaths and Recovered')
plt.xlabel('Deaths')
plt.ticklabel_format(style='plain',axis='x')
plt.ylabel('Recovered')
plt.ticklabel_format(style='plain',axis='y')
plt.tight_layout()
plt.savefig('deaths_and_recoverd.png')
plt.show()
```

**"This scatter plot compares deaths and recoveries.Counteries with more cases show both higher recoveries and deaths."**

### Distribution Of Recovery Rate (Recovery/100 Cases)

```python
plt.figure(figsize=(10,6))
plt.hist(df['Recovered / 100 Cases'],bins=50,color='green',edgecolor='black')
plt.title('Recoverd / 100  cases')
plt.xlabel('Recoverd / 100 cases')
plt.ylabel('number of countries')
plt.ticklabel_format(style='plain',axis='y')
plt.tight_layout()
plt.savefig('recovered_100_cases.png')
plt.show()
```

**"This histogram shows recovery percentage across Counteries.Most Counteries have a high recovery rate close to 100%."**

### Top 10 Countries With Highest Death Rate (Deaths/100 Cases)

```python
country_count = df.nlargest(10,'Deaths / 100 Cases')
plt.figure(figsize=(10,6))
plt.bar(country_count['Country/Region'],country_count['Deaths / 100 Cases'],color='red')
plt.title('Top 10 countries with highest deaths cases')
plt.xlabel('country')
plt.ylabel('Number of death cases')
plt.ticklabel_format(style='plain',axis='y')
plt.tight_layout()
plt.savefig('top_10_highest_deaths_cases.png')
plt.show()
```

**"This bar chart shows counteries with highest death rate (deaths per 100 cases).it highlights severityof the disease in those counteries"**
